```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean_test=100
std_dev_test=15
prob_between_90_and_110=norm.cdf(90+20,mean_test, std_dev_test)-norm.cdf(90,mean_test,std_dev_test)
prob_below_85 = norm.cdf(85, mean_test, std_dev_test)
prob_above_130 = 1 - norm.cdf(130, mean_test, std_dev_test)
print(f"Between 90 and 110: {prob_between_90_and_110:.2f}")
print(f"Below 85: {prob_below_85:.2f}")
print(f"Above 130: {prob_above_130:.2f}")
x = np.linspace(mean_test-4*std_dev_test,mean_test+4*std_dev_test)
pdf_values=norm.pdf(x,mean_test,std_dev_test)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label="Normal Distribution PDF", color='b', linewidth=2)
plt.fill_between(x, 0, pdf_values, where=((x>=90) & (x<=(90+20))), color='green', alpha=0.3, label="Between 90 and 110")
plt.fill_between(x, 0, pdf_values, where=(x <= 85), color='red', alpha=0.3, label="Below 85")
plt.fill_between(x, 0, pdf_values, where=(x >= 130), color='orange', alpha=0.3, label="Above 130")
plt.title('Normal Distribution and Calculated Probabilities')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
