```python
import pandas as pd
data = pd.read_csv('titanic_clean.csv')
summary = data.describe()
print(summary)
print(data.describe(include=['object']))
```

```python
print(data.columns)
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='skyblue')
```

```python
plt.figure(figsize=(6, 4))
data['pclass'].value_counts().plot(kind='bar', color='olive')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True,
            boxprops=dict(facecolor='pink'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sn
sn.pairplot(data[['age', 'sibsp', 'parch']].dropna())
plt.show()
```

```python
correlation_matrix = data[['age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='survived', data=data, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
